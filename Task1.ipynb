```python
def distanceBetweenPoints(vector1, vector2):
    'This function calculates/returns the distance between two vectors: vector1 and vector2'
    # Subtracting the two given vectors
        
    uv = np.subtract(vector1,vector2)

    # Finding the dot product of the 2
        
    total = np.dot(uv,uv)

    #Returning the distance between the 2 vectors
    
    return np.sqrt(total)
```

```python
def nearestNeighbour(trainingSetValues, trainingSetLabels, valueToTest):
    'This function calculates/returns the label of 1 Nearest Neighbour given: a Training dataset, its corresponding label and the value to predict a label'


    # Set the current nearest label to be 0
    nearestLabel = 0
    
    # Set the current least distance to be infinite
    leastDistance = math.inf 
    
    # For every value in the training set
    for i in range(0,trainingSetValues.shape[0]):

        # Calculate the distance between the given training set value and the value to test
        distanceBetween = distanceBetweenPoints(trainingSetValues[i], valueToTest)
    
        # If the distance calculated is less than the leastDistance then update the distance and the labels
        if distanceBetween < leastDistance:
            leastDistance = distanceBetween
            nearestLabel = trainingSetLabels[i]
    
    return nearestLabel
```

```python
def calculateSuccessRate (trainingSetValues, trainingSetLabels, testSetValues, testSetLabels):
    'This function calculates the error rate of the 1 Nearest Neighbours algorithm, given: The Training Dataset, its corresponding labels, dataset to test, corresponding labels to test data'

    # Counter variable to count the mistakes made by the program
    errorCount = 0

    predictedLabels =[]

    # For every value in the test set
    for i in range(0, len(testSetValues)):
       
       # Append the predicted labels to the new array
       predictedLabels.append(nearestNeighbour(trainingSetValues,trainingSetLabels, testSetValues[i]))

       print("Predicted Label: ", predictedLabels[i])
       print("Actual Label: ", testSetLabels[i])
       print(" ")
        
       # If the predicted does not match the correct one, add 1 to the error counter
       if predictedLabels[i] != testSetLabels[i]:
           errorCount +=1

    # Returns the Success rate in percentage
    return ((1- (errorCount/testSetLabels.shape[0])) * 100)

def calculateErrorRate(successRate):
    # Returns the error rate in percentage
    return 100 - successRate
```

```python
def neighboursWithN(trainingSetValues, trainingSetLabels, valueToTest, numberOfNeighbours):
    'This function calculates N Nearest Neighbours labels, given: The Training Dataset, its corresponding labels, value to test, number of neighbours wanted'

    nearestNeighbors = {}  # A dictionary to store nearest neighbors and distances

    # For every value in the training data set
    for i in range(trainingSetValues.shape[0]):

        # Calculate the distance between the training set value and the value we are testing
        distanceBetween = distanceBetweenPoints(trainingSetValues[i], valueToTest)

        # Check length of the dictionary to see if it is less than the number of NN wanted to add to dictionary
        # And checks if the current distance is smaller than the largest distance in the neighbors (max size in array is 4)
        if len(nearestNeighbors) < numberOfNeighbours or distanceBetween < max(nearestNeighbors.values()):

            # If it is, add this new distance to the dictionary
            nearestNeighbors[i] = distanceBetween

        # If the length of the dictionary is greater than the number of Nearest Neighbours
        if len(nearestNeighbors) > numberOfNeighbours:

            # Get the largest distance found in the dictionary
            farthest_neighbor = max(nearestNeighbors, key=nearestNeighbors.get)

            # Delete the furthest neighbour from the dictionary
            del nearestNeighbors[farthest_neighbor]



    # Get the labels of the nearest neighbors based on the i-th value stored in the dictionary
    # It will iterate through all the labels till the i-th labels are reached according to the dictionary keys
    nearestLabels = [trainingSetLabels[n] for n in nearestNeighbors.keys()]

    # Returns the array of all the nearest labels
    return nearestLabels
```

```python
def nearestNeighbourN(trainingSetValues, trainingSetLabels, valueToTest, numberOfNeighbours):
    'This function calculates N Nearest Neighbours predicted label, given: The Training Dataset, its corresponding labels, value to test, number of neighbours wanted'

    # Get all the nearest neighbours' labels
    nearest_labels = neighboursWithN(trainingSetValues, trainingSetLabels, valueToTest, numberOfNeighbours)

    # Create a dictionary with "Label: Number of occurences in array"
    label_counts = {label: nearest_labels.count(label) for label in set(nearest_labels)}   
     
    # Returns the key with the max value in the dictionary
        # list(label_counts.keys()) = Gets the all the labels from the dictionary
        # list(label_counts.values()) = Gets the values from those keys
        # .index(max(label_counts.values())) = Gets the index value of the max item
     
    return list(label_counts.keys())[list(label_counts.values()).index(max(label_counts.values()))]
```

```python
def calculateSuccessRateWithN(trainingSetValues, trainingSetLabels, testSetValues, testSetLabels, numberOfNeighbours):

    errorCount = 0
    predictedLabels =[]

    for i in range(0, len(testSetValues)):
       predictedLabels.append(nearestNeighbourN(trainingSetValues,trainingSetLabels, testSetValues[i],numberOfNeighbours))

       print("Predicted Label: ", predictedLabels[i])
       print("Actual Label: ", testSetLabels[i])
       print(" ")

       if predictedLabels[i] != testSetLabels[i]:
           errorCount +=1

    return ((1- (errorCount/testSetLabels.shape[0])) * 100)
```

```python
def nonConformityMeasure(trainingSetValues, trainingSetLabels, testValue, predictedLabel):
    nearestSameLabelD = math.inf
    nearestOtherLabelD = math.inf

    # For every item in the dataset
    for i in range(trainingSetValues.shape[0]):

        # Checks that the testValue and the training set values are not the same
        if np.array_equal(trainingSetValues[i], testValue):
            continue

        # Calculate the distance between the current item and the item to test
        distanceBetween = distanceBetweenPoints(trainingSetValues[i], testValue)

        
        # Checks if the current trainingValue has the same label as the testValue and the distance is less than previously calculated
        if trainingSetLabels[i] == predictedLabel and distanceBetween < nearestSameLabelD:
            nearestSameLabelD = distanceBetween

        # Checks if the current trainingValue does not have the same label as the testValue and the distance is less than previously calculated
        if trainingSetLabels[i] != predictedLabel and distanceBetween < nearestOtherLabelD:
            nearestOtherLabelD = distanceBetween
            

    # Returns the non-conformity value from the equation:
        # the distance to the nearest sample of a different class/ the distance to the nearest sample of the same class
    return nearestOtherLabelD/nearestSameLabelD
```

```python
def nonConformityMeasureArray(trainingSetValues, trainingSetLabels):
    'This function, given a Training Set Values and Labels, returns an array of all non-conformity scores of each training value'
    nonConformityValues =[]
    
    for i in range(0,trainingSetValues.shape[0]):
        nonConformityValues.append(nonConformityMeasure(trainingSetValues, trainingSetLabels, trainingSetValues[i], trainingSetLabels[i]))

    return nonConformityValues
```

```python
def nonConformalPrediction(trainingSetValues, trainingSetLabels, testValue):
    'This function, given the Training set values and corresponding labels as well as a value to test will return the non-conformal values for all the values'
    scores =[] 
    
    # Here I am creating a copy of the original training set as I will append the test value and all the new labels to test, iteratively to the test trainingSet
    originalTrainingSet = np.copy(trainingSetValues)
    originalTrainingSetLabels = np.copy(trainingSetLabels)


    # Adding the new value to test to the training set and changing the shape to accomodate the new value
    trainingSetValues = np.append(trainingSetValues, testValue)
    trainingSetValues = np.reshape(trainingSetValues,(originalTrainingSet.shape[0]+1,originalTrainingSet.shape[1]))

    # Count the number of labels in this training set 
    labelsToUse = countLabels(trainingSetLabels)

    # For every label, add that new testValue and a given label and calculate the non-conformity value
    # Add score to an array, clean the trainingSetLabels and move to the next label
    for i in range(0, len(labelsToUse)):
    
        trainingSetLabels = np.append(trainingSetLabels, labelsToUse.pop())    
        conformityArray = nonConformityMeasureArray(trainingSetValues, trainingSetLabels)
        scores.append(conformityArray)
        trainingSetLabels = np.copy(originalTrainingSetLabels)


    # Revert the datasets to what they were previously
    trainingSetValues = np.copy(originalTrainingSet)
    trainingSetLabels = np.copy(originalTrainingSetLabels)

    # Return an array of arrays with n dimensions where N is the number of labels
    return scores



def countLabels(trainingSetLabels):
    'This function, given a set of labels, will produce a set containing all the distinct labels'
    setLabels = set()

    for x in range(len(trainingSetLabels)-1):
        setLabels.add(trainingSetLabels[x])

    return setLabels

```

```python
def calculatePValue(trainingSetValues, trainingSetLabels, testValue, testLabel):
    'This function, given the training set values and labels, and a test value and its label, will calculate the P value for each data entry'
    pValues = []

    # Get the conformity scores arrays
        # Arrays will be conformalArrays[0] = First Label Conformity Values
    conformalArrays = nonConformalPrediction(trainingSetValues,trainingSetLabels, testValue)

    labels = countLabels(trainingSetLabels)

    for i in range(len(labels)):

        # Setting count to 1 as lowest p(y) is 1/number of values, not 0/number of values
        count = 1

        # Produce a rank for each value based on their conformity score
        sortedArray = np.array(conformalArrays[i]).argsort()

        # This for loop will calculate the rank of the value based on their position in the sortedArray
        for x in range(len(sortedArray)):

            # If value found
            if sortedArray[x] == len(conformalArrays[i])-1:
                break

            # Add 1 to counter
            else:
                count = count + 1
        

        # Calculating the P value
        pValue = (count/len(sortedArray))

        currentLabel = labels.pop()

        # This if statement, makes sure that the correct Label gives a higher P value
            # This is because the wrong label may give a higher P value but we want to make sure that we count the false P values
        if testLabel == currentLabel:
            pValues.append(pValue +1)

        else:
            pValues.append(pValue)

        print("P Value for ", testValue, " and label ", currentLabel, " is: ",pValue)


    return pValues
```

```python
def calculateAvgFalseP(trainingSetValues, trainingSetLabels, testSetValues, testSetLabels):
    'This function, given the training set values and labels, the test values and their labels, will calculate average false P value for the datasets'
    falsePValues= []

    for i in range(0, testSetValues.shape[0]):

        # Calculates the P value for a given value and label
        pValues = calculatePValue(trainingSetValues, trainingSetLabels, testSetValues[i],testSetLabels[i])
        print(" ")
        
        # Removes the max value, i.e the correct label P value
        maxP = max(pValues)
        pValues.remove(maxP)

        # Appends the false values to an array
        falsePValues.append(pValues)

    cumSum = 0
    counter = 0

    # For every value in the false P values array, sum all the values 
    for x in range(len(falsePValues)):
        for z in range(len(falsePValues[x])):
            cumSum =  cumSum + falsePValues[x][z]
            counter = counter + 1

    # Return average false P-Value
    return cumSum/counter
```
